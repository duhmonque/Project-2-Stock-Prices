# Import Dependencies, Assign Variables, Assign Stock ID

```python
#Dependencies
import pandas as pd

#Files to Load for Stock Prices
google_stock_path = "./google.csv"
yahoo_stock_path = "./yahoo.csv"

#Read the data file
google_stock = pd.read_csv(google_stock_path)
google_stock['stock_ID'] = 1
yahoo_stock = pd.read_csv(yahoo_stock_path)
yahoo_stock['stock_ID'] = 2
```

# Date as String

```python
#
google_stock['Date'][0]
```

# Formatting Date to Date Time

```python
#New column
google_stock['Date'] = pd.to_datetime(google_stock['Date'])
```

# Renaming Column Names, Align Column Names Between CSVs

```python
google_stock = google_stock.rename(columns={"Date": "date", "High": "high","Low": "low", "Open": "open", 
                             "Close": "close", "Volume": "volume","Adj Close": "adj_close",
                             "stock_ID": "stock_id"})
google_stock
```

# Identifying Data Types and Column Name Updates

```python
google_stock.dtypes
```

# Complete Same Process for Second CSV (Yahoo)

```python
#
yahoo_stock['Date'][0]
```

```python
yahoo_stock['Date'] = pd.to_datetime(yahoo_stock['Date'])
yahoo_stock = yahoo_stock.rename(columns={"Date": "date", "High": "high","Low": "low", "Open": "open", 
                             "Close": "close", "Volume": "volume","Adj Close": "adj_close",
                             "stock_ID": "stock_id"})
yahoo_stock
```

```python
yahoo_stock.dtypes
```

# Concat/Combine Two Tables into One DF

```python
# Display the data table for preview
complete_stock_data = pd.concat([google_stock, yahoo_stock])
complete_stock_data.reset_index(inplace = True, drop = True)
```
